```python
# Import the os and subprocess libraries
import os
import subprocess
```

```python
# Create a docker file in your project directory (next to your Python Notebook)
# The content of the file should be this:

# FROM ubuntu
```

```python
# This function prints the contents of the docker file
# If the message: No Dockerfile found in the current directory
# make sure docker is installed in your computer and a docker
# file is present in the directory

# FROM ubuntu

def get_docker_file_info():
    print("\nDockerfile Information\n")
    # check if Dockerfile exists
    if os.path.exists("Dockerfile"):
        # display contents of Dockerfile
        with open("Dockerfile", "r") as file:
            print(file.read())
    else:
        print("No Dockerfile found in the current directory.")

get_docker_file_info()
```

```python
# This function runs docker commands

def execute_command(command):
    print("\nRunning Containers\n")
    # run "docker ps" command
    result = subprocess.run(command, capture_output=True)
    # display output of command
    print(result.stdout.decode())
```

```python
# Modify this function so that it creates an image based on the dockerfile

command = ["docker", "build", "-t", "test", "."]
execute_command(command)
```

```python
# Modify this function so that it gets information about the image from the Docker file

execute_command(["docker", "images", "test"])
```
